```python
import math
import matplotlib.pyplot as plt

# Додаємо магічне слово для відображення графіків у Colab
%matplotlib inline

def draw_pifagoras_tree(branch_length, level, angle=45, x=0, y=0):
    """
    Рекурсивно малює фрактал "Дерево Піфагора".

    Аргументи:
        branch_length: float - Довжина гілки.
        level: int - Рівень рекурсії.
        angle: float - Кут повороту.
        x, y: float - Початкові координати (за замовчуванням: початок координат).

    Повертає:
        None
    """
    if level == 0:
        return

    # Відображення гілки дерева
    x1, y1 = x, y
    x2 = x1 + branch_length * math.cos(math.radians(angle))
    y2 = y1 + branch_length * math.sin(math.radians(angle))
    plt.plot([x1, x2], [y1, y2], color='black')

    # Виклик функції для лівої гілки
    draw_pifagoras_tree(branch_length / 2, level - 1, angle, x2, y2)
    # Виклик функції для правої гілки
    draw_pifagoras_tree(branch_length / 2, level - 1, -angle, x2, y2)

def main():
    # Введення рівня рекурсії від користувача
    level = int(input("Введіть рівень рекурсії: "))
    branch_length = 30
    angle = 15

    # Встановлення розмірів графіку
    plt.figure(figsize=(8, 8))
    # Встановлення видимих меж графіку
    plt.axis('equal')
    plt.axis('off')

    # Виклик рекурсивної функції для малювання дерева Піфагора
    draw_pifagoras_tree(branch_length, level)

    # Показати графік
    plt.show()

if __name__ == "__main__":
    main()
```
